Regular Expression allows us to search for specific patterns in the text.

```python
import re
```

```python
text = '''Mr. John Smith email john@gmail.com, john@stanford.edu smith@nasa.mit.edu
Mrs. Jessy Smith contact number 111-234-9876 , 123.456.7890 , 025*004*1991
Ms Jennet has website https://jennet.com, jennet.com
Mr Jacob websites are https://www.jacob-world.net
Hotline Numbers 800-123-1234 , 900-321-4321 , 700-678-6789
hot pot dot shotr
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
0123456789
Characters that needs to be scaped (as they have special operations to perform):
. \ & ^ ( ) { } [ ] * + |
hehe he'''
```

```python
with open('./../data/50-contacts.csv') as infile:
    data = infile.read()
```

```python
pattern =re.compile(r'abc')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
pattern =re.compile(r'hoisa')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
pattern =re.compile(r'.')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`.` - Any Character except new line

```python
pattern =re.compile(r'\.')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
pattern =re.compile(r'jennet\.com')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\d` - matched Digits (0-9)

```python
pattern =re.compile(r'\d')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\D` - matches **NOT** Digits

```python
pattern =re.compile(r'\D')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\w` - matches Word Characters (a-z A-Z 0-9 _)

```python
pattern =re.compile(r'\w')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\W` - matched **NOT** a Word Character

```python
pattern =re.compile(r'\W')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\s` - Whitespaces (space, tab, newline)

```python
pattern =re.compile(r'\s')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\S` - Matches **NOT** Whitespaces

```python
pattern =re.compile(r'\S')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\b<pattern>` -  Word Boundary

```python
pattern =re.compile(r'\bhe')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`\B<pattern>` - **Not** a Word Boundary

```python
pattern =re.compile(r'\Bhe')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`^` - Beggining of a String

```python
pattern =re.compile(r'^Mr')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
pattern =re.compile(r'^The')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`$` - End of a String

```python
pattern =re.compile(r'he$')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
pattern =re.compile(r'the$')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`[0-5]` - Matches in range (a-z, A-Z, 0-9)

```python
pattern =re.compile(r'[0-6]')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`[^0-5]` - **NOT** Matches in range. `^` inside `[]` used for negation

```python
pattern =re.compile(r'[^a-zA-Z]')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

```python
pattern =re.compile(r'[^d]ot')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

did not match dot

`<pattern>*` - Matches 0 or more

```python
pattern =re.compile(r'\w*[^d]ot')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`<pattern>+` - mathces 1 or more

`<pattern>?` - Matches 0 or 1

```python
pattern =re.compile(r'Mr\.?\s[A-Z]\w*')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

`<pattern>{count}` - Matches exact number of pattern

`<pattern>{min-max}` - Matches range of numbers of pattern

`|` - Either or
`( )` - Group

```python
pattern =re.compile(r'M(r|s|rs)\.?\s[A-Z]\w*')
matches = pattern.finditer(text)
for match in matches:
    print(match)
```

#### Finding Website

```python
pattern =re.compile(r'https?://(www\.)?([\w-]+)(\.\w+)')
matches = pattern.finditer(text)
for i, match in enumerate(matches, start=1):
    print(i, match, '---', match.group(0), '|', match.group(1), '|', match.group(2), '|', match.group(3))
```

```python
pattern =re.compile(r'https?(www\.)?\S+')
matches = pattern.finditer(text)
for i, match in enumerate(matches, start=1):
    print(i, match)
```

### findall() method

`findall() method` finds all the string matching the pattern. `finditer()` methods also does the same but it gives some extra information such as span of the string.

```python
pattern =re.compile(r'[a-zA-Z0-9._-]+@[a-zA-Z-]+\.[a-zA-Z-.]+')
matches = pattern.findall(data)
for i, match in enumerate(matches, start=1):
    print(i, match)
```

### search() method

`search()` returns the first match object if the pattern exist else it returns `None`. It does not return any iterable. It just looks for the first available matching pattern.

```python
pattern =re.compile(r'[a-zA-Z0-9._-]+@[a-zA-Z-]+\.[a-zA-Z-.]+')
matched = pattern.search(data)
print(matched)
```

```python
pattern =re.compile(r'\d{20}')
matched = pattern.search(data)
print(matched)
```
